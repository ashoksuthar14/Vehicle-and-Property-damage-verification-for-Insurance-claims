```python
# Import necessary libraries
import pandas as pd
import numpy as np
import networkx as nx
import torch
import torch.nn.functional as F
from torch_geometric.data import Data, HeteroData
from torch_geometric.nn import HeteroConv, GCNConv, SAGEConv, GATConv
from torch_geometric.transforms import ToUndirected
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Clean and preprocess the data
def preprocess_data(df):
    # Handle missing values
    df['authorities_contacted'].fillna('None', inplace=True)
    df['police_report_available'].replace('?', 'UNKNOWN', inplace=True)

    # Convert date columns to datetime
    df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
    df['incident_date'] = pd.to_datetime(df['incident_date'])

    # Extract useful features from dates
    df['policy_age_days'] = (df['incident_date'] - df['policy_bind_date']).dt.days

    # Convert target variable to binary
    df['fraud_binary'] = df['fraud_reported'].map({'Y': 1, 'N': 0})

    return df

# Apply preprocessing
df_cleaned = preprocess_data(df)

# Display fraud distribution
fraud_count = df_cleaned['fraud_reported'].value_counts()
print("\nFraud Distribution:")
print(fraud_count)
print(f"Fraud Rate: {fraud_count['Y'] / len(df_cleaned) * 100:.2f}%")
```

```python
# Step 4: Feature Engineering for Graph Construction

# 1. Encode categorical features
def encode_features(df):
    # Initialize encoders
    label_encoders = {}
    categorical_cols = ['policy_state', 'policy_csl', 'insured_sex', 'insured_education_level',
                         'insured_occupation', 'insured_hobbies', 'insured_relationship',
                         'incident_type', 'collision_type', 'incident_severity',
                         'authorities_contacted', 'incident_state', 'incident_city',
                         'property_damage', 'police_report_available', 'auto_make', 'auto_model']

    # Apply label encoding
    df_encoded = df.copy()
    for col in categorical_cols:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # Scale numerical features
    numerical_cols = ['months_as_customer', 'age', 'policy_deductable', 'policy_annual_premium',
                       'umbrella_limit', 'capital-gains', 'capital-loss', 'incident_hour_of_the_day',
                       'number_of_vehicles_involved', 'bodily_injuries', 'witnesses',
                       'total_claim_amount', 'injury_claim', 'property_claim', 'vehicle_claim',
                       'auto_year', 'policy_age_days']

    scaler = StandardScaler()
    df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

    return df_encoded, label_encoders

# 2. Create node features
def create_node_features(df_encoded):
    # Customer features (policyholder)
    customer_features = df_encoded[['age', 'insured_sex', 'insured_education_level',
                                     'insured_occupation', 'insured_hobbies', 'insured_relationship',
                                     'capital-gains', 'capital-loss', 'months_as_customer']].values

    # Policy features
    policy_features = df_encoded[['policy_deductable', 'policy_annual_premium',
                                   'umbrella_limit', 'policy_state', 'policy_csl',
                                   'policy_age_days']].values

    # Claim features
    claim_features = df_encoded[['incident_type', 'collision_type', 'incident_severity',
                                 'authorities_contacted', 'incident_state', 'incident_city',
                                 'incident_hour_of_the_day', 'number_of_vehicles_involved',
                                 'property_damage', 'bodily_injuries', 'witnesses',
                                 'police_report_available', 'total_claim_amount',
                                 'injury_claim', 'property_claim', 'vehicle_claim']].values

    # Vehicle features
    vehicle_features = df_encoded[['auto_make', 'auto_model', 'auto_year']].values

    return customer_features, policy_features, claim_features, vehicle_features

# Apply encoding
df_encoded, label_encoders = encode_features(df_cleaned)

# Create node features
customer_features, policy_features, claim_features, vehicle_features = create_node_features(df_encoded)

# Convert to PyTorch tensors
customer_x = torch.tensor(customer_features, dtype=torch.float)
policy_x = torch.tensor(policy_features, dtype=torch.float)
claim_x = torch.tensor(claim_features, dtype=torch.float)
vehicle_x = torch.tensor(vehicle_features, dtype=torch.float)

# Print shapes to verify
print(f"Customer features shape: {customer_x.shape}")
print(f"Policy features shape: {policy_x.shape}")
print(f"Claim features shape: {claim_x.shape}")
print(f"Vehicle features shape: {vehicle_x.shape}")
```

```python
# Step 5: Build a Heterogeneous Graph

# Create a heterogeneous graph data object
def build_heterogeneous_graph(df, customer_x, policy_x, claim_x, vehicle_x):
    data = HeteroData()

    # Add node features for each type
    data['customer'].x = customer_x
    data['policy'].x = policy_x
    data['claim'].x = claim_x
    data['vehicle'].x = vehicle_x

    # Add node labels (fraud or not)
    data['claim'].y = torch.tensor(df['fraud_binary'].values, dtype=torch.long)

    # Create node index maps for policy, customer, and vehicle
    # Using policy_number, customer_id (we'll use index as proxy), and auto combination as unique identifiers
    n_nodes = len(df)
    policy_mapping = {policy_id: idx for idx, policy_id in enumerate(df['policy_number'])}
    customer_mapping = {idx: idx for idx in range(n_nodes)}  # Customer ID = index in this case
    vehicle_mapping = {idx: idx for idx in range(n_nodes)}  # Vehicle ID = index in this case

    # Create edges between different node types
    # Customer to Policy edges
    customer_to_policy_src = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    customer_to_policy_dst = torch.tensor([policy_mapping[pid] for pid in df['policy_number']], dtype=torch.long)
    data['customer', 'has_policy', 'policy'].edge_index = torch.stack([customer_to_policy_src, customer_to_policy_dst])

    # Policy to Claim edges
    policy_to_claim_src = torch.tensor([policy_mapping[pid] for pid in df['policy_number']], dtype=torch.long)
    policy_to_claim_dst = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    data['policy', 'has_claim', 'claim'].edge_index = torch.stack([policy_to_claim_src, policy_to_claim_dst])

    # Customer to Claim edges
    customer_to_claim_src = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    customer_to_claim_dst = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    data['customer', 'filed_claim', 'claim'].edge_index = torch.stack([customer_to_claim_src, customer_to_claim_dst])

    # Vehicle to Claim edges
    vehicle_to_claim_src = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    vehicle_to_claim_dst = torch.tensor(list(range(n_nodes)), dtype=torch.long)
    data['vehicle', 'involved_in', 'claim'].edge_index = torch.stack([vehicle_to_claim_src, vehicle_to_claim_dst])

    # Create additional edges based on incident data
    # Claims with same incident location
    location_groups = df.groupby('incident_location').indices
    for location, indices in location_groups.items():
        if len(indices) > 1:  # Only create edges if multiple claims share location
            for i in range(len(indices)):
                for j in range(i+1, len(indices)):
                    # Add bidirectional edges
                    src = torch.tensor([indices[i], indices[j]], dtype=torch.long)
                    dst = torch.tensor([indices[j], indices[i]], dtype=torch.long)

                    # Concatenate with existing edges if they exist
                    if ('claim', 'same_location', 'claim') in data.edge_types:
                        existing_src, existing_dst = data['claim', 'same_location', 'claim'].edge_index
                        data['claim', 'same_location', 'claim'].edge_index = torch.cat([
                            torch.stack([existing_src, existing_dst]),
                            torch.stack([src, dst])
                        ], dim=1)
                    else:
                        data['claim', 'same_location', 'claim'].edge_index = torch.stack([src, dst])

    # Add edges between claims with similar characteristics
    # For example, claims with same incident type
    incident_groups = df.groupby('incident_type').indices
    for incident_type, indices in incident_groups.items():
        if len(indices) > 1:
            for i in range(len(indices)):
                for j in range(i+1, len(indices)):
                    # Add bidirectional edges
                    src = torch.tensor([indices[i], indices[j]], dtype=torch.long)
                    dst = torch.tensor([indices[j], indices[i]], dtype=torch.long)

                    # Concatenate with existing edges if they exist
                    if ('claim', 'similar_incident', 'claim') in data.edge_types:
                        existing_src, existing_dst = data['claim', 'similar_incident', 'claim'].edge_index
                        data['claim', 'similar_incident', 'claim'].edge_index = torch.cat([
                            torch.stack([existing_src, existing_dst]),
                            torch.stack([src, dst])
                        ], dim=1)
                    else:
                        data['claim', 'similar_incident', 'claim'].edge_index = torch.stack([src, dst])

    # Create train, validation, and test masks
    num_claims = data['claim'].x.size(0)
    perm = torch.randperm(num_claims)

    train_idx = perm[:int(0.7 * num_claims)]
    val_idx = perm[int(0.7 * num_claims):int(0.85 * num_claims)]
    test_idx = perm[int(0.85 * num_claims):]

    # Binary masks
    train_mask = torch.zeros(num_claims, dtype=torch.bool)
    val_mask = torch.zeros(num_claims, dtype=torch.bool)
    test_mask = torch.zeros(num_claims, dtype=torch.bool)

    train_mask[train_idx] = True
    val_mask[val_idx] = True
    test_mask[test_idx] = True

    data['claim'].train_mask = train_mask
    data['claim'].val_mask = val_mask
    data['claim'].test_mask = test_mask

    return data

# Build the heterogeneous graph
hetero_data = build_heterogeneous_graph(df_cleaned, customer_x, policy_x, claim_x, vehicle_x)

# Print graph statistics
print("\nHeterogeneous Graph Statistics:")
print(f"Number of node types: {len(hetero_data.node_types)}")
print(f"Node types: {hetero_data.node_types}")
print(f"Number of edge types: {len(hetero_data.edge_types)}")
print(f"Edge types: {hetero_data.edge_types}")

for node_type in hetero_data.node_types:
    print(f"{node_type} nodes: {hetero_data[node_type].x.shape[0]}")

for edge_type in hetero_data.edge_types:
    if hetero_data[edge_type].edge_index.numel() > 0:
        print(f"{edge_type} edges: {hetero_data[edge_type].edge_index.shape[1]}")
```

```python
class HeteroGNN(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels, metadata):
        super().__init__()

        # Layer 1: Different convolutional layers for different edge types
        self.conv1 = HeteroConv({
            ('customer', 'has_policy', 'policy'): GATConv((-1, -1), hidden_channels, heads=3, add_self_loops=False),
            ('policy', 'has_claim', 'claim'): GATConv((-1, -1), hidden_channels, heads=3, add_self_loops=False),
            ('customer', 'filed_claim', 'claim'): GATConv((-1, -1), hidden_channels, heads=3, add_self_loops=False),
            ('vehicle', 'involved_in', 'claim'): GATConv((-1, -1), hidden_channels, heads=3, add_self_loops=False),
            # SAGEConv doesn't accept add_self_loops parameter
            ('claim', 'similar_incident', 'claim'): SAGEConv((-1, -1), hidden_channels)
        }, aggr='sum')

        # Layer 2
        self.conv2 = HeteroConv({
            ('customer', 'has_policy', 'policy'): GATConv((-1, -1), hidden_channels, heads=2, add_self_loops=False),
            ('policy', 'has_claim', 'claim'): GATConv((-1, -1), hidden_channels, heads=2, add_self_loops=False),
            ('customer', 'filed_claim', 'claim'): GATConv((-1, -1), hidden_channels, heads=2, add_self_loops=False),
            ('vehicle', 'involved_in', 'claim'): GATConv((-1, -1), hidden_channels, heads=2, add_self_loops=False),
            # SAGEConv doesn't accept add_self_loops parameter
            ('claim', 'similar_incident', 'claim'): SAGEConv((-1, -1), hidden_channels)
        }, aggr='sum')

        # Final classifier for fraud detection on claim nodes
        self.lin = torch.nn.Linear(hidden_channels, out_channels)

        # Attention mechanism for feature importance
        self.attention = torch.nn.Linear(hidden_channels, 1)

        # Batch normalization layers
        self.batch_norm1 = torch.nn.ModuleDict({
            node_type: torch.nn.BatchNorm1d(hidden_channels * 3 if node_type in ['customer', 'policy', 'claim', 'vehicle'] else hidden_channels)
            for node_type in metadata[0]
        })

        self.batch_norm2 = torch.nn.ModuleDict({
            node_type: torch.nn.BatchNorm1d(hidden_channels * 2 if node_type in ['customer', 'policy', 'claim', 'vehicle'] else hidden_channels)
            for node_type in metadata[0]
        })

        # Dropout
        self.dropout = torch.nn.Dropout(0.3)

    def forward(self, x_dict, edge_indices_dict):
        # First heterogeneous graph convolution
        x_dict = self.conv1(x_dict, edge_indices_dict)

        # Apply non-linearity and batch normalization
        x_dict = {key: self.batch_norm1[key](F.relu(x)) for key, x in x_dict.items()}
        x_dict = {key: self.dropout(x) for key, x in x_dict.items()}

        # Second heterogeneous graph convolution
        x_dict = self.conv2(x_dict, edge_indices_dict)

        # Apply non-linearity and batch normalization
        x_dict = {key: self.batch_norm2[key](F.relu(x)) for key, x in x_dict.items()}
        x_dict = {key: self.dropout(x) for key, x in x_dict.items()}

        # Apply attention to claim node features - only for 'claim' nodes
        claim_features = x_dict['claim']
        attention_weights = F.softmax(self.attention(claim_features), dim=0)
        x_dict['claim'] = attention_weights * claim_features

        # Apply final linear layer to claim nodes
        out = self.lin(x_dict['claim'])

        return out, attention_weights

# Initialize and train the model
hidden_channels = 64
out_channels = 2  # Binary classification (fraud or not)
model = HeteroGNN(hidden_channels, out_channels, hetero_data.metadata())
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

# Train the model
print("\nTraining the GNN model...")
model, train_losses, val_losses, val_accs, val_aucs = train_model(
    model, hetero_data, optimizer, criterion, epochs=100)
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import RGCNConv, Linear

class InsuranceRGCN(torch.nn.Module):
    def __init__(self, node_features, hidden_channels, out_channels, num_relations, num_node_types):
        super().__init__()

        # R-GCN layers - specifically designed for heterogeneous graphs
        self.conv1 = RGCNConv(node_features, hidden_channels, num_relations)
        self.conv2 = RGCNConv(hidden_channels, hidden_channels, num_relations)

        # Node type embeddings - help differentiate between node types
        self.node_type_embedding = torch.nn.Embedding(num_node_types, hidden_channels)

        # Output layers for different tasks
        self.claim_classifier = torch.nn.Sequential(
            Linear(hidden_channels, hidden_channels),
            torch.nn.ReLU(),
            torch.nn.Dropout(0.3),
            Linear(hidden_channels, out_channels)
        )

        # Batch normalization
        self.batch_norm1 = torch.nn.BatchNorm1d(hidden_channels)
        self.batch_norm2 = torch.nn.BatchNorm1d(hidden_channels)

        # Dropout for regularization
        self.dropout = torch.nn.Dropout(0.3)

        # Initialize parameters
        self.reset_parameters()

    def reset_parameters(self):
        self.conv1.reset_parameters()
        self.conv2.reset_parameters()
        torch.nn.init.xavier_uniform_(self.node_type_embedding.weight)

    def forward(self, data):
        # Prepare node features and edge information for R-GCN
        x = data.x_dict

        # Preprocess features and put all nodes in a single tensor
        # First convert heterogeneous features to homogeneous
        node_types, edge_types = {}, {}
        node_features = []
        edge_indices = []
        edge_relations = []

        # Assign type indices to node types
        for i, node_type in enumerate(data.node_types):
            node_types[node_type] = i

        # Assign relation indices to edge types
        for i, edge_type in enumerate(data.edge_types):
            edge_types[edge_type] = i

        # Collect node features
        num_nodes = 0
        node_type_indices = []
        node_offsets = {}

        # Create homogeneous node feature tensor
        for node_type, features in x.items():
            node_offsets[node_type] = num_nodes
            num_nodes_of_type = features.size(0)
            node_features.append(features)
            node_type_indices.extend([node_types[node_type]] * num_nodes_of_type)
            num_nodes += num_nodes_of_type

        # Concatenate all node features
        if all(isinstance(feat, torch.Tensor) for feat in node_features):
            # Ensure all feature tensors have the same number of columns
            # If not, pad with zeros
            max_cols = max(feat.size(1) for feat in node_features)
            padded_features = []
            for feat in node_features:
                if feat.size(1) < max_cols:
                    padding = torch.zeros(feat.size(0), max_cols - feat.size(1), device=feat.device)
                    padded_features.append(torch.cat([feat, padding], dim=1))
                else:
                    padded_features.append(feat)
            x = torch.cat(padded_features, dim=0)
        else:
            # Handle the case where features might not be tensors
            x = torch.zeros(num_nodes, data.num_features)

        # Get node type embeddings
        node_type_tensor = torch.tensor(node_type_indices, device=x.device)
        node_type_emb = self.node_type_embedding(node_type_tensor)

        # Create homogeneous edge index tensor and edge relation tensor
        for edge_type, edge_index in data.edge_index_dict.items():
            src_type, relation, dst_type = edge_type
            src_offset = node_offsets[src_type]
            dst_offset = node_offsets[dst_type]

            # Adjust node indices with offsets
            adjusted_edge_index = edge_index.clone()
            adjusted_edge_index[0] += src_offset
            adjusted_edge_index[1] += dst_offset

            edge_indices.append(adjusted_edge_index)
            edge_relations.extend([edge_types[edge_type]] * edge_index.size(1))

        # Concatenate all edge indices and create relation tensor
        edge_index = torch.cat(edge_indices, dim=1)
        edge_relation = torch.tensor(edge_relations, device=edge_index.device)

        # First R-GCN layer
        x = self.conv1(x, edge_index, edge_relation)
        x = self.batch_norm1(x)
        x = F.relu(x)
        x = self.dropout(x)

        # Enhance features with node type embeddings
        x = x + node_type_emb

        # Second R-GCN layer
        x = self.conv2(x, edge_index, edge_relation)
        x = self.batch_norm2(x)
        x = F.relu(x)
        x = self.dropout(x)

        # For each node type, extract corresponding features
        x_dict = {}
        start_idx = 0
        for node_type, features in data.x_dict.items():
            num_nodes = features.size(0)
            x_dict[node_type] = x[start_idx:start_idx + num_nodes]
            start_idx += num_nodes

        # Apply classification only for claim nodes
        if 'claim' in x_dict:
            claim_out = self.claim_classifier(x_dict['claim'])
            # Generate attention weights for feature importance
            attention_weights = F.softmax(torch.abs(x_dict['claim']).mean(dim=1, keepdim=True), dim=0)
            return claim_out, attention_weights
        else:
            # Fallback if no claim nodes
            return None, None

# Create the model
def create_rgcn_model(data, hidden_channels=64, out_channels=2):
    # Count number of relations and node types
    num_relations = len(data.edge_types)
    num_node_types = len(data.node_types)

    # Get the feature dimension (assuming all node types have the same feature dimension)
    # If different, use the maximum
    node_features = max([feat.size(1) for feat in data.x_dict.values()])

    model = InsuranceRGCN(node_features, hidden_channels, out_channels,
                        num_relations, num_node_types)

    return model

# Initialize the model
hidden_channels = 64
out_channels = 2
model = create_rgcn_model(hetero_data, hidden_channels, out_channels)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

# Train the model
print("\nTraining the R-GCN model...")
# You'll need to modify your train_model function to accept the data as a whole
# rather than x_dict and edge_indices_dict separately
model, train_losses, val_losses, val_accs, val_aucs = train_model(
    model, hetero_data, optimizer, criterion, epochs=100)
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.data import HeteroData
from torch_geometric.nn import HeteroConv, SAGEConv

###############################################################################
# 1) DUMMY DATA CREATION (Replace with your real HeteroData construction)
###############################################################################
hetero_data = HeteroData()

num_claims = 1000
num_customers = 1000
num_policies = 1000
num_vehicles = 1000

claim_features = 16
customer_features = 9
policy_features = 6
vehicle_features = 3

# Create x features
hetero_data['claim'].x = torch.randn(num_claims, claim_features)
hetero_data['customer'].x = torch.randn(num_customers, customer_features)
hetero_data['policy'].x = torch.randn(num_policies, policy_features)
hetero_data['vehicle'].x = torch.randn(num_vehicles, vehicle_features)

# Example bipartite edges (use random for demo)
# In a real scenario, you must ensure the node indices are correct:
#   i.e., edge_index for ('customer','has_policy','policy') should
#   only contain valid indices in [0, num_customers) for the source
#   and [0, num_policies) for the target.
hetero_data['customer','has_policy','policy'].edge_index = torch.randint(
    0, min(num_customers,num_policies), (2, 5000))
hetero_data['policy','has_claim','claim'].edge_index = torch.randint(
    0, min(num_policies,num_claims), (2, 5000))
hetero_data['customer','filed_claim','claim'].edge_index = torch.randint(
    0, min(num_customers,num_claims), (2, 5000))
hetero_data['vehicle','involved_in','claim'].edge_index = torch.randint(
    0, min(num_vehicles,num_claims), (2, 5000))

# Two homogeneous edges on 'claim' node type
hetero_data['claim','similar_incident','claim'].edge_index = torch.randint(
    0, num_claims, (2, 3000))
hetero_data['claim','same_location','claim'].edge_index = torch.randint(
    0, num_claims, (2, 3000))

# Labels, train/val/test masks on 'claim' nodes
hetero_data['claim'].y = torch.randint(0, 2, (num_claims,))
train_mask = torch.zeros(num_claims, dtype=torch.bool)
val_mask = torch.zeros(num_claims, dtype=torch.bool)
test_mask = torch.zeros(num_claims, dtype=torch.bool)
train_mask[:700] = True
val_mask[700:850] = True
test_mask[850:] = True
hetero_data['claim'].train_mask = train_mask
hetero_data['claim'].val_mask = val_mask
hetero_data['claim'].test_mask = test_mask

###############################################################################
# 2) DEFINE YOUR FRAUD HETEROGNN MODEL
###############################################################################
class FraudHeteroGNN(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels):
        super().__init__()

        #######################################################################
        # First HeteroConv:
        #   - Bipartite edges: SAGEConv((-1, -1), hidden_channels)
        #   - Homogeneous edges: SAGEConv(-1, hidden_channels)
        #######################################################################
        self.conv1 = HeteroConv({
            # Bipartite edges:
            ('customer', 'has_policy', 'policy'):  SAGEConv((-1, -1), hidden_channels),
            ('policy',   'has_claim', 'claim'):    SAGEConv((-1, -1), hidden_channels),
            ('customer', 'filed_claim', 'claim'):  SAGEConv((-1, -1), hidden_channels),
            ('vehicle',  'involved_in','claim'):   SAGEConv((-1, -1), hidden_channels),

            # Homogeneous edges on 'claim':
            ('claim', 'similar_incident', 'claim'): SAGEConv(-1, hidden_channels),
            ('claim', 'same_location', 'claim'):    SAGEConv(-1, hidden_channels),
        }, aggr='sum')

        #######################################################################
        # Second HeteroConv:
        #   - Bipartite edges: (hidden, hidden)
        #   - Homogeneous edges: single hidden
        #######################################################################
        self.conv2 = HeteroConv({
            ('customer', 'has_policy', 'policy'):  SAGEConv((hidden_channels, hidden_channels), hidden_channels),
            ('policy',   'has_claim', 'claim'):    SAGEConv((hidden_channels, hidden_channels), hidden_channels),
            ('customer', 'filed_claim', 'claim'):  SAGEConv((hidden_channels, hidden_channels), hidden_channels),
            ('vehicle',  'involved_in','claim'):   SAGEConv((hidden_channels, hidden_channels), hidden_channels),

            ('claim', 'similar_incident', 'claim'): SAGEConv(hidden_channels, hidden_channels),
            ('claim', 'same_location', 'claim'):    SAGEConv(hidden_channels, hidden_channels),
        }, aggr='sum')

        # Final linear layer for 'claim' node classification
        self.lin = torch.nn.Linear(hidden_channels, out_channels)

    def forward(self, x_dict, edge_index_dict):
        # Pass through the first HeteroConv
        x_dict = self.conv1(x_dict, edge_index_dict)
        for node_type in x_dict:
            x_dict[node_type] = F.relu(x_dict[node_type])

        # Pass through the second HeteroConv
        x_dict = self.conv2(x_dict, edge_index_dict)
        for node_type in x_dict:
            x_dict[node_type] = F.relu(x_dict[node_type])

        # Output from 'claim' node embeddings
        return self.lin(x_dict['claim']), x_dict

###############################################################################
# 3) TRAINING & EVALUATION
###############################################################################
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = FraudHeteroGNN(hidden_channels=32, out_channels=2).to(device)
hetero_data = hetero_data.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

def train():
    model.train()
    optimizer.zero_grad()
    out, _ = model(hetero_data.x_dict, hetero_data.edge_index_dict)
    train_mask = hetero_data['claim'].train_mask
    loss = criterion(out[train_mask], hetero_data['claim'].y[train_mask])
    loss.backward()
    optimizer.step()
    return float(loss)

@torch.no_grad()
def evaluate(split='val'):
    model.eval()
    out, _ = model(hetero_data.x_dict, hetero_data.edge_index_dict)

    if split == 'train':
        mask = hetero_data['claim'].train_mask
    elif split == 'val':
        mask = hetero_data['claim'].val_mask
    else:
        mask = hetero_data['claim'].test_mask

    logits = out[mask]
    preds = logits.argmax(dim=1)
    labels = hetero_data['claim'].y[mask]
    acc = (preds == labels).sum() / mask.sum()
    return float(acc), preds, labels

# Run training
for epoch in range(1, 11):
    loss_val = train()
    val_acc, _, _ = evaluate('val')
    if epoch % 2 == 0:
        print(f"Epoch {epoch} | Loss: {loss_val:.4f} | Val Acc: {val_acc:.4f}")

test_acc, test_preds, test_labels = evaluate('test')
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_meta_graph(hetero_data):
    """
    Plots a meta-graph of your HeteroData object.
    Each node in the plot is a node type.
    Each edge in the plot is a relation type (with label).
    """
    G = nx.DiGraph()

    # 1) Add each node type as a node in our meta-graph
    for node_type in hetero_data.node_types:
        G.add_node(node_type)

    # 2) Add directed edges for each (src_type, relation, dst_type)
    for (src_type, relation, dst_type) in hetero_data.edge_types:
        # If you prefer an undirected meta-graph, use nx.Graph() above,
        # or add both directions manually for each relation.
        G.add_edge(src_type, dst_type, label=relation)

    # 3) Layout and draw
    pos = nx.spring_layout(G, k=1.0)  # You can tweak 'k' for spacing
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue")

    # Draw the edge labels (relation names)
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.title("Meta-Graph of Node Types and Relations")
    plt.show()

# Example usage:
plot_meta_graph(hetero_data)
```

```python
import random
import networkx as nx
import matplotlib.pyplot as plt

def plot_fraud_subgraph(hetero_data, max_fraud=5, max_non_fraud=5):
    """
    Visualize a small subgraph showing a few fraud and non-fraud 'claim' nodes,
    plus their neighbors. Fraud claims are colored differently from non-fraud
    claims. Other node types (e.g. 'customer', 'policy', etc.) have their own color.

    Args:
        hetero_data (HeteroData): Your heterogeneous graph data.
        max_fraud (int): Number of fraud claims to sample.
        max_non_fraud (int): Number of non-fraud claims to sample.
    """
    # 1) Identify fraud vs. non-fraud claim indices
    claim_labels = hetero_data['claim'].y
    fraud_idx = (claim_labels == 1).nonzero(as_tuple=True)[0].tolist()
    non_fraud_idx = (claim_labels == 0).nonzero(as_tuple=True)[0].tolist()

    # If you have fewer fraud or non-fraud claims than requested, adjust accordingly
    num_fraud = min(len(fraud_idx), max_fraud)
    num_non_fraud = min(len(non_fraud_idx), max_non_fraud)

    # 2) Randomly sample from each category
    random.shuffle(fraud_idx)
    random.shuffle(non_fraud_idx)
    fraud_idx = fraud_idx[:num_fraud]
    non_fraud_idx = non_fraud_idx[:num_non_fraud]

    # 3) Collect these claim nodes and their neighbors
    # We'll gather the edges for each edge type that involves 'claim' as src or dst.
    # Then keep only edges connected to our chosen claims (fraud or non-fraud).

    chosen_claims = set(fraud_idx + non_fraud_idx)
    node_set = set(chosen_claims)  # We'll keep track of node IDs for each node type

    # For building a NetworkX multi-type graph, we'll store edges as:
    # (('node_type', node_id), ('node_type', node_id))
    # This disambiguates node_type A's index 10 from node_type B's index 10.
    edges_to_add = []

    # Let's iterate over all edge types in hetero_data:
    for edge_type in hetero_data.edge_types:
        src_type, relation, dst_type = edge_type
        edge_index = hetero_data[edge_type].edge_index

        # We'll see which side is 'claim', so we know which indices to filter.
        # CASE A: claim is the source
        if src_type == 'claim':
            # src are claim indices, dst are "other" node type
            src_list = edge_index[0].tolist()
            dst_list = edge_index[1].tolist()

            # Keep only edges where the claim src is in our chosen_claims
            filtered_edges = []
            for s, d in zip(src_list, dst_list):
                if s in chosen_claims:
                    # keep this edge
                    filtered_edges.append((s, d))
                    node_set.add(d)  # neighbor node (in the other node type)

            # Add edges to our list in a way that captures node types
            for (s, d) in filtered_edges:
                edges_to_add.append(
                    (('claim', s), (dst_type, d), relation)  # store relation if you want
                )

        # CASE B: claim is the destination
        elif dst_type == 'claim':
            # src are "other" type, dst are claim
            src_list = edge_index[0].tolist()
            dst_list = edge_index[1].tolist()

            filtered_edges = []
            for s, d in zip(src_list, dst_list):
                if d in chosen_claims:
                    filtered_edges.append((s, d))
                    node_set.add(s)

            for (s, d) in filtered_edges:
                edges_to_add.append(
                    ((src_type, s), ('claim', d), relation)
                )

        # CASE C: claim -> claim (homogeneous edges)
        elif src_type == 'claim' and dst_type == 'claim':
            # For edges among claims themselves (similar_incident, same_location, etc.)
            src_list = edge_index[0].tolist()
            dst_list = edge_index[1].tolist()

            filtered_edges = []
            for s, d in zip(src_list, dst_list):
                # Keep edge if s or d is in chosen_claims (some people prefer both s and d in chosen_claims)
                # We'll keep it if either is in chosen_claims to see cross-connections
                if s in chosen_claims or d in chosen_claims:
                    filtered_edges.append((s, d))
                    # add them to the node set
                    node_set.add(s)
                    node_set.add(d)

            for (s, d) in filtered_edges:
                edges_to_add.append(
                    (('claim', s), ('claim', d), relation)
                )

    # 4) Build a NetworkX graph from these edges
    G = nx.Graph()  # or nx.DiGraph if you prefer direction

    # Add nodes with attributes: 'node_type', 'is_fraud' for claim
    # We stored everything in node_set only for claim node IDs.
    # But we also might have neighbors from 'customer', 'policy', etc. so let's do a pass:
    # Actually, we have them in edges_to_add as well.

    # edges_to_add is a list of ((node_type, node_id), (node_type, node_id), relation)
    # We'll add the nodes from the edges themselves.
    for (u, v, rel) in edges_to_add:
        # u => (u_type, u_id)
        # v => (v_type, v_id)
        G.add_node(u, node_type=u[0])
        G.add_node(v, node_type=v[0])
        # Optional: store the relation as an edge attribute
        G.add_edge(u, v, relation=rel)

    # Mark fraud vs non-fraud for 'claim' nodes
    # Because we have up to claim_id = num_claims
    # we can check if claim_id is in chosen_claims, and if it's in fraud_idx
    for (n_type, n_id) in G.nodes():
        if n_type == 'claim':
            if n_id in fraud_idx:
                G.nodes[(n_type, n_id)]['fraud'] = True
            else:
                G.nodes[(n_type, n_id)]['fraud'] = False
        else:
            G.nodes[(n_type, n_id)]['fraud'] = None  # Not applicable

    # 5) Assign colors: let's do:
    # - Fraud claim: red
    # - Non-fraud claim: green
    # - Other node types: lightblue
    color_map = []
    for node in G.nodes():
        n_type = G.nodes[node]['node_type']
        is_fraud = G.nodes[node]['fraud']
        if n_type == 'claim':
            if is_fraud is True:
                color_map.append("red")
            else:
                color_map.append("green")
        else:
            color_map.append("lightblue")

    # 6) Draw the subgraph
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, k=0.9, seed=42)  # tweak 'k' for spacing
    nx.draw_networkx_nodes(G, pos, node_color=color_map, node_size=500)
    nx.draw_networkx_labels(
        G, pos,
        labels={n: f"{n[0]}-{n[1]}" for n in G.nodes()},  # e.g. "claim-12", "policy-7"
        font_size=8
    )
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.7)

    # Optionally draw edge labels (the relation type):
    edge_labels = {(u, v): G[u][v]['relation'] for (u, v) in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7)

    plt.title("Subgraph of Fraud vs Non-Fraud Claims and Neighbors")
    plt.axis("off")
    plt.show()

# Example usage (assuming you have a HeteroData named 'hetero_data'):
# plot_fraud_subgraph(hetero_data, max_fraud=5, max_non_fraud=5)
```
