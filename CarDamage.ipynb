```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Load the pre-trained YOLOv8 model
model = YOLO('/content/best (4).pt')  # Replace with your trained YOLOv8 model path
```

```python
# Helper function to load an image (in Colab, use files.upload() to upload)
def load_image(image_path):
    img = Image.open(image_path)
    img = img.convert('RGB')  # Ensure image is in RGB
    return np.array(img)
```

```python
# Function to detect damages
def detect_damages(image):
    results = model(image)
    damage_info = []

    for detection in results[0].boxes:
        class_id = int(detection.cls.cpu().numpy())  # Class label ID
        confidence = float(detection.conf.cpu().numpy())  # Confidence score
        bbox = detection.xyxy.cpu().numpy()  # Bounding box coordinates (x_min, y_min, x_max, y_max)

        # Append damage details to list
        damage_info.append({
            "class_id": class_id,
            "confidence": confidence,
            "bbox": bbox
        })

    return results, damage_info
```

```python
# Function to visualize detections
def visualize_damages(image, results, damage_info, class_names):
    for damage in damage_info:
        bbox = damage['bbox']
        class_id = damage['class_id']
        confidence = damage['confidence']

        # Draw bounding boxes and labels on the image
        # The problem likely lies here where bbox might have an unexpected shape.
        # We will add a check and flatten the array if needed.

        # Check if bbox is 2D and flatten if needed
        if bbox.ndim == 2:
            bbox = bbox.flatten()

        x_min, y_min, x_max, y_max = map(int, bbox)
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)
        cv2.putText(image, f"{class_names[class_id]} ({confidence:.2f})", (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
# Define cost ranges for each damage type (example)
damage_cost_map = {
    0: ('crack_and_hole', 100, 300),
    1: ('medium_deformation', 200, 500),
    2: ('severe_deformation', 500, 1000),
    3: ('severe_scratch', 100, 300),
    4: ('slight_deformation', 50, 150),
    5: ('slight_scratch', 20, 80),
    6: ('windshield_damage', 200, 400)
}

# Function to calculate the estimated cost
def estimate_cost(damage_info):
    total_cost = 0
    cost_breakdown = []

    for damage in damage_info:
        class_id = damage['class_id']
        confidence = damage['confidence']
        min_cost, max_cost = damage_cost_map[class_id][1:3]

        # Scale cost within range based on confidence
        estimated_cost = min_cost + (max_cost - min_cost) * confidence
        total_cost += estimated_cost

        cost_breakdown.append({
            "type": damage_cost_map[class_id][0],
            "estimated_cost": estimated_cost
        })

    return total_cost, cost_breakdown
```

```python
# Complete function for full pipeline
def car_damage_pipeline(image_path):
    # Load and process image
    image = load_image(image_path)
    image_copy = image.copy()

    # Detect damages
    results, damage_info = detect_damages(image)

    # Visualize damages
    class_names = model.names  # Get class names from YOLOv8 model
    visualize_damages(image_copy, results, damage_info, class_names)

    # Estimate cost
    total_cost, cost_breakdown = estimate_cost(damage_info)

    # Display cost breakdown
    print("\n--- Estimated Repair Costs ---")
    for cost in cost_breakdown:
        print(f"{cost['type'].capitalize()}: ${cost['estimated_cost']:.2f}")
    print(f"\nTotal Estimated Repair Cost: ${total_cost:.2f}")
```
